```python
import matplotlib.pyplot as plt

def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz_sequence(n):
    sequence = collatz_sequence(n)
    x = list(range(1, len(sequence) + 1))

    plt.plot(x, sequence, marker='o', linestyle='-')

    # Add y-value labels on each step
    for i, j in zip(x, sequence):
        plt.text(i, j, str(j), ha='right', va='bottom')

    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.title(f'Collatz Sequence for n = {n}')
    plt.grid(True)
    plt.show()

# Input a positive integer
try:
    n = int(input("Enter a positive integer: "))
    if n <= 0:
        print("Please enter a positive integer.")
    else:
        plot_collatz_sequence(n)
except ValueError:
    print("Invalid input. Please enter a positive integer.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz_sequence(n):
    sequence = collatz_sequence(n)
    x = list(range(1, len(sequence) + 1))

    # Apply a logarithmic transformation to the y-values
    y = np.log10(sequence)

    plt.plot(x, y, marker='o', linestyle='-')

    # Add y-value labels on each step
    for i, j in zip(x, y):
        plt.text(i, j, f'{sequence[i-1]}', ha='right', va='bottom')

    plt.xlabel('Step')
    plt.ylabel('log10(Value)')
    plt.title(f'Collatz Sequence for n = {n} (logarithmic scale)')
    plt.grid(True)
    plt.show()

# Input a positive integer
try:
    n = int(input("Enter a positive integer: "))
    if n <= 0:
        print("Please enter a positive integer.")
    else:
        plot_collatz_sequence(n)
except ValueError:
    print("Invalid input. Please enter a positive integer.")
```
